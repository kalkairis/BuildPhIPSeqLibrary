```python
frequency_per_aa = from_web_table.groupby('amino_acid').frequency_per_thousand.sum().drop(index='*').to_frame()
frequency_per_aa['frequency_per_thousand'] /= frequency_per_aa['frequency_per_thousand'].sum()

num_options_per_aa = from_web_table.groupby('amino_acid').codon.count()
```

```python
res = {}
for barcode_aa_length in range(1, 10):
    rounds = []
    for round_i in range(1000):
        if round_i%10==0:
            print(f"In round {round_i}")
        res = {}
        for i in range(1000000):
            sequence = np.random.choice(frequency_per_aa.index, size=barcode_aa_length, 
                                        p=frequency_per_aa['frequency_per_thousand'])
            num_options = res.get(''.join(sequence), np.prod(list(map(num_options_per_aa.get, sequence))))
            if num_options == 0:
                break
            res[''.join(sequence)] = num_options - 1
        rounds.append(i)
    pd.Series(rounds).to_csv(f"round_for_{barcode_aa_length}.csv")
    res[barcode_aa_length] = rounds
```
